```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
from IPython.display import Audio

%matplotlib inline
```

### Download Multichannel .wav file

Download audio file *6_Channel_ID.wav* from http://www-mmsp.ece.mcgill.ca/Documents/AudioFormats/WAVE/Samples/Microsoft/8_Channel_ID.wav

### Read file and split channels

```python
Fs, data = read('6_Channel_ID.wav')
print("Sampling Frequency is", Fs)
```

```python
FL = data[:,0]
FR = data[:,1]
FC = data[:,2]
LF = data[:,3]
BL = data[:,4]
BR = data[:,5]
```

### Play audio in Notebook

```python
Audio(FL, rate=Fs)
```

```python
Audio(FR, rate=Fs)
```

```python
Audio(FC, rate=Fs)
```

```python
Audio(LF, rate=Fs)
```

```python
Audio(BL, rate=Fs)
```

```python
Audio(BR, rate=Fs)
```

### Plot the Wave form

```python
plt.figure()
plt.plot(data)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Waveform of Test Audio')
plt.show()
```
